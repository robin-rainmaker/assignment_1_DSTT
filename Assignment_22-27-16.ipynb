```python
# question 05 / code 
#NAME : ROBIN SINGH RAUTHAN | Reg .no: 22-27-16  |  Branch : Data science | Department : SoCs&Ds
num=int(input("enter the no of element to find  factorial "))
def factorial(n):
  if (n==1 or n==0):
    return 1
  else:
    return (n*factorial(n-1))

result= factorial(num)
print(f"factorial {num}! =" , result)
```
